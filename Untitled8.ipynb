```python
# -----------------------------
# Q5 â€” Tokenization
# -----------------------------

import re
import spacy

# Load English tokenizer
nlp = spacy.load("en_core_web_sm")

paragraph = """
I love studying NLP. It's challenging, but it's also exciting!
Sometimes, tokenization behaves strangely.
"""

# 1. Naive space-based tokenization
naive_tokens = paragraph.split()
print("Naive tokens:", naive_tokens)

# 2. Manual correction
manual_tokens = re.findall(r"[A-Za-z']+", paragraph)
print("Manual tokens:", manual_tokens)

# 3. spaCy tokenization
doc = nlp(paragraph)
spacy_tokens = [t.text for t in doc]
print("spaCy tokens:", spacy_tokens)

# Compare differences
print("\nDifferences:")
print("Tokens in spaCy but not manual:", set(spacy_tokens) - set(manual_tokens))
print("Tokens in manual but not spaCy:", set(manual_tokens) - set(spacy_tokens))
```
